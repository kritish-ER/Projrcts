```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("synthetic_employee_dataset.csv")

# 1. Display the first 10 rows of the dataset
print("First 10 rows of the dataset:")
print(df.head(10))

# 2. Show complete dataset information including data types
print("\nDataset Info:")
print(df.info())

# 3. Determine and display the number of columns and rows
print("\nShape of the dataset:")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

# 4. Univariate Analysis

# a) Descriptive statistics for numerical columns
print("\nDescriptive Statistics (Numerical Columns):")
print(df.describe())

# b) Value counts for categorical variables
print("\nValue Counts for Categorical Variables:")
categorical_columns = df.select_dtypes(include='object').columns
for col in categorical_columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())

# c) Distribution plots for numerical variables
print("\nPlotting Distribution of Numerical Variables:")
numerical_columns = df.select_dtypes(include='number').columns
for col in numerical_columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

# d) Frequency analysis (value counts) of all columns
print("\nFrequency Analysis of All Columns:")
for col in df.columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())
```
